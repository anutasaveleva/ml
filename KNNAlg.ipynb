```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def euclideanDist(x1,y1,x2,y2):
    return math.sqrt((x1-y2)**2+(y1-y2)**2)
```

```python
def knn(test_data, train_data, k):
    clusters = []
    for i in range(len(test_data)):
        distances = []
        for j in range(len(train_data)):
            dist=euclideanDist(test_data[i][0], test_data[i][1], train_data[j][0][0], train_data[j][0][1])
            distances.append([dist, train_data[j][1]])
        idx = sorted(distances)[:k]
        clusters_c = dict.fromkeys([l for l in range(clust_num)], 0)
        
        for j in range(len(idx)):
            clusters_c[idx[j][1]] += 1
            
        max_v = max(clusters_c[j] for j in range(0, clust_num))
        max_c = 0
        
        for clust, total in clusters_c.items():
            if total == max_v:
                max_c = clust
        clusters.append(max_c)
    return clusters
```

```python
def generate_data(point_num, classnum):
    dataset = []
    for i in range(classnum):
        x_coord, y_coord = random.random()*5.0, random.random()*5.0
        for j in range(point_num):
            dataset.append([[random.gauss(x_coord,0.5), random.gauss(y_coord,0.5)], i])
    return dataset
```

```python
n = 300
clust_num = 3
data = generate_data(n,clust_num)

test_data = [data[i][0] for i in range(len(data)) if i % 2 == 0]
train_data = [data[i] for i in range(len(data)) if i % 2 != 0]

colors = ['r','g','blue','y','black']
for i in (train_data):
    plt.scatter(i[0][0], i[0][1],color=colors[i[1]])
```

```python
k = 5
clusters = knn(test_data, train_data, k)
for i in range(len(test_data)):
    plt.scatter(test_data[i][0], test_data[i][1], c=colors[clusters[i]])
```
